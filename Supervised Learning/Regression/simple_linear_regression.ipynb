## Simple Linear Regression from scratch

Basic assumption of a simple linear regression is that there is a linear relationship between in the input feature $x$ and the output $y$

```python
import pandas as pd 
import numpy as np 

import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
x = np.random.normal(0, 1, 100)
```

```python
beta_0 = 1 
beta_1 = 2

e = np.random.normal(0, 0.1, 100)

y_hat = beta_0 + beta_1 * x + e 
```

```python
plt.plot(x, y_hat, 'bo')
plt.plot(x, 1+2*x, 'b-')
```

### Loss Function (MSE)

$$J(\beta_0, \beta_1) =  \dfrac 1 n \sum_{i=1}^n (y_i - \hat y_i)^2$$

For $\beta_0$ (intercept):

$$ \frac {\partial J}{\partial  \beta_0} = -\frac{2}{n} \sum_{i=1}^{n}(y_i-(\beta_0 + \beta_1 x_i))$$

For $\beta_1$ (intercept):

$$ \frac {\partial J}{\partial  \beta_1} = -\frac{2}{n} \sum_{i=1}^{n} x_i(y_i-(\beta_0 + \beta_1 x_i))$$



```python
beta_0 = 0
beta_1 = 0

alpha = 0.01
num_iterations = 1000
```

```python
for i in range(num_iterations):
    
    y_pred = beta_0 + beta_1 * x 
    
    d_beta_0 = (-2 / len(x)) * sum(y_hat - y_pred)
    d_beta_1 = (-2 / len(x)) * sum(x * (y_hat - y_pred))

    beta_0 = beta_0 - alpha * d_beta_0
    beta_1 = beta_1 - alpha * d_beta_1
    
```

```python
final_predictions =  beta_0 + beta_1 * x
mse = np.mean((y_hat - final_predictions)**2)
```

```python
plt.scatter(x, y_hat, color='blue', label='Data points')
plt.plot(x, final_predictions, color='red', label='Regression line')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
